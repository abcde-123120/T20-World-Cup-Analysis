T20 WORLD CUP 2022 ANALYSIS USING PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("t20-world-cup-22.csv")
data
```

Number of matches won by each team in the world cup:

```python
data['winner'].value_counts().plot(kind='bar')
plt. title('Number of Matches Won by teams in t20 world Cup 2022')
plt.xticks(rotation=45)
plt.show()
```

Number of matches won by batting first or second in the t20 world cup 2022:

```python
won_by=  data['won by'].value_counts()
label = won_by.index
counts = won_by.values
colors = ['pink','purple']

plt.pie(counts,labels = label,colors=colors,autopct='%1.2f%%')
plt.title('Number of Matches Won By Runs or Wickets')
plt.show()
```

TOSS DECISIONS IN T20 WORLD CUP 2022

```python
toss = data['toss decision'].value_counts()
label=toss.index
counts=toss.values
colors=['skyblue','orange']
plt.pie(counts,labels=label,colors=colors,autopct='%1.2f%%')
plt.title('Toss Decision in T20 World Cup 2022')
plt.show()
```

TOP SCORERS IN THE T20 WORLD CUP 2022:

```python
data_grouped=data.groupby('top scorer')['highest score'].sum().reset_index()
plt.bar(data_grouped['top scorer'],data_grouped['highest score'],color='blue')
plt.title('Top scorers in t20 world cup 2022')
plt.xticks(rotation = 90)
plt.xlabel('top scorer')
plt.ylabel('highest score')
plt.show()
```

Number of Player of the Match  awards in the world cup

```python
player_of_match_counts=data['player of the match'].value_counts()
plt.bar(player_of_match_counts.index,player_of_match_counts.values)
plt.title("Player of the Match Awards in t20 World Cup 2022")
plt.xlabel('player of the match')
plt.ylabel('number of awards')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
